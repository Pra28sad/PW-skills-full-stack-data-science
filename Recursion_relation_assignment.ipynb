### 3. Given a recurrence relation, solve it using the recursive tree approach:
### a. T(n) = 2T(n-1) +1
### b. T(n) = 2T(n/2) + n

##### a. T(n) = 2T(n-1) +1

In this recurrence relation, at each step, the function calls itself twice with 
n1. This means that at each level of recursion, there are two branches, each 
representing a call to T(n1). The additional "+1" represents some constant amount 
of work done at each level.

##### b. T(n) = 2T(n/2) + n

In this recurrence relation, at each step, the function divides n by 2 and makes two recursive calls. Each of these calls is on a problem of size n/2, and then there's an additional linear amount of work,n.
